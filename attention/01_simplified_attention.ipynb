# Simplified Self-Attention

Attention without trainable weights. Each token attends to every other token via raw dot products between embeddings.

```python
import torch

inputs = torch.tensor(
  [[0.43, 0.15, 0.89], # Your     (x^1)
   [0.55, 0.87, 0.66], # journey  (x^2)
   [0.57, 0.85, 0.64], # starts   (x^3)
   [0.22, 0.58, 0.33], # with     (x^4)
   [0.77, 0.25, 0.10], # one      (x^5)
   [0.05, 0.80, 0.55]] # step     (x^6)
)
```

## Step 1: Attention scores for a single query token

Pick token 2 ("journey") as the query. Compute its dot product with every other token.
High score = that token is more relevant to the query.

```python
query = inputs[1]  # x^2 — "journey"

attn_scores_2 = torch.empty(inputs.shape[0])
for i, x_i in enumerate(inputs):
    attn_scores_2[i] = torch.dot(x_i, query)

print(attn_scores_2)
```

## Step 2: Normalize with softmax

Softmax converts raw scores into probabilities that sum to 1.
These become the attention weights — how much each token contributes to the context vector.

```python
attn_weights_2 = torch.softmax(attn_scores_2, dim=0)
print("Attention weights:", attn_weights_2)
print("Sum:", attn_weights_2.sum())
```

## Step 3: Context vector

Weighted sum of all input embeddings using attention weights.
This is the new representation of token 2 — enriched with information from all other tokens.

```python
context_vec_2 = attn_weights_2 @ inputs
print("Context vector:", context_vec_2)
```

## Step 4: Compute attention for ALL tokens at once

Instead of looping over each query, use matrix multiplication to get all attention scores in one shot.
Each row i contains the attention scores of token i against all other tokens.

```python
attn_scores = inputs @ inputs.T
print("Attention scores (all pairs):\n", attn_scores)
```

```python
attn_weights = torch.softmax(attn_scores, dim=-1)
print("Attention weights:\n", attn_weights)

context_vecs = attn_weights @ inputs
print("\nContext vectors (all tokens):\n", context_vecs)
```
